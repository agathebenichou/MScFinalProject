```python
def generate_fhr_sample(patient_data, duration=10, resolution=1):

    num_points = int(duration / resolution)
    min_fhr = patient_data['histogram_min']
    max_fhr = patient_data['histogram_max']
    mode_fhr = patient_data['histogram_mode']
    mean_fhr = patient_data['histogram_mean']
    median_fhr = patient_data['histogram_median']
    variance = patient_data['histogram_variance']
    peaks = int(patient_data['histogram_number_of_peaks'])
    zeroes = int(patient_data['histogram_number_of_zeroes'])
    tendency = patient_data['histogram_tendency']

    # Generating normal distribution around mean with given variance
    normal_dist = np.random.normal(loc=mean_fhr, scale=np.sqrt(variance), size=num_points)

    # Clipping values to stay within min and max FHR
    fhr_sample = np.clip(normal_dist, min_fhr, max_fhr)

    # Introducing peaks and zeroes if applicable
    if peaks > 0 or zeroes > 0:
        peak_indices = np.random.choice(range(num_points), peaks + zeroes, replace=False)
        for i in peak_indices:
            if zeroes > 0:
                fhr_sample[i] = 0
                zeroes -= 1
            else:
                fhr_sample[i] = np.random.choice([min_fhr, max_fhr])

    # Adjusting for tendency
    if tendency != 0:
        trend = np.linspace(0, tendency * duration, num_points)
        fhr_sample += trend

    return np.linspace(0, duration, num_points), fhr_sample

random_patient_data = data.sample(1).iloc[0]

time_points, fhr_sample = generate_fhr_sample(random_patient_data)

plt.figure(figsize=(12, 6))
plt.plot(time_points, fhr_sample, color='green')
plt.title('Generated Sample FHR Data over 10 Minutes')
plt.xlabel('Time (minutes)')
plt.ylabel('FHR (beats per minute)')
plt.grid(True)
plt.show()
```

```python
def generate_fhr_samples_for_health_categories(dataset, health_categories, duration=10, resolution=1):
    fhr_samples = {}
    for health in health_categories:

        filtered_data = dataset[dataset['fetal_health'] == health]
        random_patient_data = filtered_data.sample(1).iloc[0]

        time_points, fhr_sample = generate_fhr_sample(patient_data=random_patient_data, duration=30, resolution=resolution)
        fhr_samples[health] = (time_points, fhr_sample)

    return fhr_samples

# Generating FHR samples for fetuses with health 1, 2, and 3
health_category_labels = {1: 'Normal', 2: 'Suspect', 3: 'Pathological'}
fhr_samples = generate_fhr_samples_for_health_categories(data, health_category_labels)

# Plotting the generated FHR samples for each health category
plt.figure(figsize=(12, 6))
for health in health_category_labels:
    time_points, fhr_sample = fhr_samples[health]
    plt.plot(time_points, fhr_sample, label=f'{health_category_labels[health]} Health')

plt.title('Generated Sample FHR Data over 10 Minutes for Different Health Categories')
plt.xlabel('Time (minutes)')
plt.ylabel('FHR (beats per minute)')
plt.legend()
plt.grid(True)
plt.show()

# split into 3 graphs on same xy axes
# use histogram columns to augment data, especially for suspect and pathological samples
# modify graph to 10 minute intervals with 1 min ticks
```
